
#### Additional Dataframe Operations

16. Perform a `merge` using "client_list.csv" and "customer_id_list.csv". Assign the resulting dataframe as `clients`.
17. Perform a `merge` using `clients` and "purchase_log.csv" and limit the subset to only clients who made purchases. Assign the resulting dataframe as `detailed_sales`.
18. Use `groupby` to find the client who spent the most money on purchases. Determine how much he/she spent. HINT: save the intermediate dataframe from using `groupby` as `spenders` before applying slicing to determine the client who spent the most money on purchases.
19. (BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.

```python
#16
df1=pd.read_csv("client_list.csv")
df1.columns=df1.columns.str.upper()
df1.head()
```

```python
df2=pd.read_csv("customer_id_list.csv")
df2.head()
```

```python
clients=pd.merge(df1,df2[["CUSTOMER_ID"]], left_index=True, right_index=True)
clients
```

```python
#17
#Perform a merge using clients and "purchase_log.csv" and limit the subset to only clients who made purchases. 
#Assign the resulting dataframe as detailed_sales.
df3=pd.read_csv("purchase_log.csv")
df3.head()
```

```python
df3.info()
```

```python
detailed_sales=pd.merge(clients,df3, on="CUSTOMER_ID")
detailed_sales
```

```python
#18
#Use groupby to find the client who spent the most money on purchases. Determine how much he/she spent. 
#HINT: save the intermediate dataframe from using groupby as spenders before applying slicing to determine the client who spent the most money on purchases.
spenders=detailed_sales
spenders=detailed_sales.groupby('CUSTOMER_ID')['PRICE'].sum().reset_index()
spenders
```

```python
max_spenders = spenders[spenders["PRICE"] == spenders["PRICE"].max()]
max_spenders
```

```python
#19
#(BONUS) Modify the answer to problem 3 slightly to determine exactly what items where purchased by the top spending client.
max_spenders_costumid=max_spenders["CUSTOMER_ID"].values[0]
```

```python
detailed_sales[detailed_sales["CUSTOMER_ID"] == max_spenders_costumid]
```
